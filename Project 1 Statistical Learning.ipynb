```python
import pandas as pd
```

# Read the Data Set and Replace the Dash with Zeros

```python
# Read data file in df variable
df=pd.read_csv('Laliga.csv',skiprows=1)
print (df)
```

```python
# show column name
df.columns
```

```python
# List of columns having - which has to be replaced with 0
columns_with_dash=['Seasons', 'Points', 'GamesPlayed', 'GamesWon',
       'GamesDrawn', 'GamesLost', 'GoalsFor', 'GoalsAgainst', 'Champion',
       'Runner-up', 'Third', 'Fourth', 'Fifth', 'Sixth', 'T']
```

```python
# Replace '-' with '0'
upd_cols_with_zero=df[columns_with_dash].replace('-',0)
```

```python
# Over write existing column of df data frame with dataframe having non dash values
df[columns_with_dash]=upd_cols_with_zero

# Check for the correctness of the data after replacing with Zero
print(df)
```

```python
df.columns
```

```python
df_debut_year=pd.DataFrame(df.Debut.str.split('-',1).tolist(),columns=['Debut_Year','Debut_cont_Year'])
print(df_debut_year)

```

```python
df_debut_year.Debut_cont_Year=df_debut_year.Debut_Year.str[:2]+df_debut_year.Debut_cont_Year
```

```python
df_debut_year=df_debut_year.fillna(0)
```

```python
df=df.join(df_debut_year)
```

```python
df_last_app=pd.DataFrame(df['Since/LastApp'].str.split('-',1).tolist(),columns=['LastApp_Year','LastApp_cont_Year'])
```

```python
df_last_app.LastApp_cont_Year=df_last_app.LastApp_Year.str[:2]+df_last_app.LastApp_cont_Year
```

```python
df_last_app=df_last_app.fillna(0)
```

```python
df=df.apply(pd.to_numeric, errors='ignore')
```

# Goal differences

```python
def func(x):
    goal_diff_count=x.GoalsFor-x.GoalsAgainst
    return goal_diff_count
```

```python
# Team goal difference
goal_diff=func(df)
print(goal_diff)
```

```python
df['GoalDiff']=goal_diff
```

```python
max_goal_diff=df.nlargest(1,'GoalDiff')
```

```python
# Team with max goal diff
max_goal_diff.Team
```

```python
min_goal_diff=df.nsmallest(1,'GoalDiff')
```

```python
# Team with min goal difference
min_goal_diff.Team
```
